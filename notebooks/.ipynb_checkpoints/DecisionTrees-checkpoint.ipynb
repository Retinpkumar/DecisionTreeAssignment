```python
# Separating positively and negatively skewed features

positively_skewed = [feat for feat in continuousFeatures if df_cont[feat].skew()>0.7]
negatively_skewed = [feat for feat in continuousFeatures if df_cont[feat].skew()<-0.7]
```

```python
print("Positively skewed features: ", positively_skewed)
print("\nNegatively skewed features: ",negatively_skewed)
```

```python
# Checking for outliers

skewed_features = positively_skewed + negatively_skewed
outlier_dict={}

for feat in continuousFeatures:
    if feat in skewed_features:
        # Applying IQR for identifying outliers for skewed features
        q1 = df_cont[feat].quantile(0.25)
        q3 = df_cont[feat].quantile(0.75)
        iqr = q3 - q1
        upper = q3+(1.5*iqr)
        lower = q1-(1.5*iqr)
        outlier_count = np.sum(df_cont[feat]>upper)+np.sum(df_cont[feat]<lower)
    else:
        # Applying 2 standard deviations for normally distributed features
        feat_mean = np.mean(df_cont[feat])
        feat_stdev = np.std(df_cont[feat])
        upper = feat_mean+(3*feat_stdev)
        lower = feat_mean-(3*feat_stdev)
        outlier_count = np.sum(df_cont[feat]>upper)+np.sum(df_cont[feat]<lower)
    outlier_dict[feat]=outlier_count
    
for feat,count in outlier_dict.items():
    print("Outliers in {}: {}%".format(feat, (count*100/df.shape[0]).round(2)))
```

## Data Preprocessing and Model evaluation

```python
# Function for dropping duplicates
def drop_duplicates(df):
    df = df.drop_duplicates()
    return df
```

```python
# Splitting the dataset
data = drop_duplicates(data)
x = data.drop(columns=['Survived'])
y = data['Survived']
```

```python
# Combining sibsp and parch to create a new feature 'Family'
def create_family(df):
    df['Family'] = df['SibSp'] + df['Parch']
    df = df.drop(columns=['SibSp','Parch'], axis=1)
    # Binarizing Family
    df['Family'] = np.where(df['Family']==0,0,1)
    return df
```

```python
# Impute missing values
def impute_nan(df):
    df['Age'].fillna(df['Age'].median(), inplace=True)
    return df
```

```python
# Binarizing Sex
def binarize(df):
    df['Sex'] = np.where(df['Sex']=='male', 1, 0)
    return df
```

```python
# Function for clipping outliers
def clip_outliers(df, cols):
    for feat in cols:
        if abs(df[feat].skew()) > 0.7:
            # Applying IQR for identifying outliers for skewed features
            q1 = df[feat].quantile(0.25)
            q3 = df[feat].quantile(0.75)
            iqr = q3 - q1
            upper = q3+(1.5*iqr)
            lower = q1-(1.5*iqr)
            df[feat] = np.where(df[feat]>upper, upper, np.where(df[feat]<lower, lower, df[feat]))
        else:
            # Applying 2 standard deviations for normally distributed features
            feat_mean = np.mean(df_cont[feat])
            feat_stdev = np.std(df_cont[feat])
            upper = feat_mean+(3*feat_stdev)
            lower = feat_mean-(3*feat_stdev)
            df[feat] = np.where(df[feat]>upper, upper, np.where(df[feat]<lower, lower, df[feat]))
    return df
```

```python
# Function for one hot encoding
def ohe(df, cols):
    for col in cols:
        df_ohe = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat([df, df_ohe], axis=1)
        df = df.drop(columns=[col], axis=1)
    return df
```

```python
x_trf = (x.
         pipe(impute_nan).
         pipe(create_family).
         pipe(binarize).
         pipe(clip_outliers, ['Age']).
         pipe(ohe, ['Pclass'])
        )
```

```python
x_trf.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

x_train, x_test, y_train, y_test = train_test_split(x_trf, y, test_size=0.2, random_state=42 )
```

```python
import graphviz
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()

decision_tree = decision_tree.fit(x_train, y_train)
```

```python
y_pred = decision_tree.predict(x_test)
```

```python
# Plotting the decision tree

import graphviz

feature_names = ['Sex', 'Age', 'Fare', 'Family','Pclass_2','Pclass_3']

class_names = ['0','1']

# DOT data
dot_data = tree.export_graphviz(decision_tree, out_file=None, 
                                feature_names=feature_names,  
                                class_names=class_names,
                                rounded=True,
                                filled=True)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
graph.render("decision_tree_graphivz")
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
conf_mat
```

```python
# Calculating positives and negatives

true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_negative = conf_mat[1][1]
```

```python
# Accuracy
Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)
Accuracy
```

```python
# Precison
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
# Recall
Recall = true_positive/(true_positive+false_negative)
Recall
```

```python
# F1 Score
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
F1_Score
```

```python
# Area Under Curve
auc = roc_auc_score(y_test, y_pred)
auc
```

```python
# ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
```

```python
# Plotting ROC
fig = plt.subplots(figsize=(5,5))
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--',label='ROC curve (area = %0.2f)' % auc)
plt.xlabel('False Positive Rate',fontsize=15)
plt.ylabel('True Positive Rate', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=18)
plt.legend(fontsize=12)
plt.show()
```
